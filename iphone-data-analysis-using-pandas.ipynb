```python
df=pd.read_csv('/kaggle/input/iphone-dataset/iphone.csv')
df
```

```python
#The column names have spaces . rename the column names to have underscore '_' instead of space 
#(try to do in one go instead of specifying each column nam in rename method)
#df.columns=['Product_Name', 'Product_URL', 'Brand', 'Sale_Price', 'Mrp','Number_Of_Ratings', 'Number_Of_Reviews', 'Upc', 'Star_Rating', 'Ram']
df.columns=df.columns.str.replace(' ','_')
df
```

```python
#Star_rating for some of the models is missing in the dataset. 
#fill those missing values with the average rating all the models.
avg_star=df['Star_Rating'].mean()
df['Star_Rating']=df['Star_Rating'].fillna(avg_star)
df
```

```python
#Now instead of filling missing values with avg rating of full dataset ,
#fill with avg rating based on RAM. 
#example :  if rating for a 2 gb phone is missing then take average of all 
#other 2 gb phones rating and fill that value.

avg_ram=df.groupby('Ram')['Star_Rating'].mean()
df['Star_Rating']=df['Star_Rating'].fillna(avg_ram)
df
```

```python
#create a new column in the dataframe "Discount_Percentage" based on MRP and sale value
df['Discount_Percentage']=round((df['Mrp']-df['Sale_Price'])*100/df['Mrp'],0)
df
```

```python
#which model has highest percent discount ?
fc=(df['Discount_Percentage']==df['Discount_Percentage'].max())
df[fc]
#df[df['Discount_Percentage']==df['Discount_Percentage'].max()]
```

```python
#find total no of reviews for iphone 11 and iphone 12 category . 
#Output should have only 2 rows (for 11 and 12)
iphone_11=df[df['Product_Name'].str.contains('iphone 11',case=False)]
iphone_12=df[df['Product_Name'].str.contains('iphone 12',case=False)]
iphone_11_count=iphone_11['Number_Of_Reviews'].sum()
iphone_12_count=iphone_12['Number_Of_Reviews'].sum()

total_reviews=pd.DataFrame({
    'iPhone_Category':['iPhone_11','iPhone_12'],
    'Review_count':[iphone_11_count,iphone_12_count]
})
total_reviews
```
